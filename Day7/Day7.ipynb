```python
#file = open('example.txt', 'r')
file = open('input.txt', 'r')

lines = file.readlines()
```

```python
# read the directory strtucture

cur_dir = '/home'
directories = {}

for line in lines:
    line=line.strip()
    if '$ cd' in line:
        if '$ cd ..' in line:
            cur_dir = cur_dir.rpartition('/')[0]
        else:
            if line.rpartition(' ')[-1] != '/':
                cur_dir = cur_dir + '/' + line.rpartition(' ')[-1]
    elif '$ ls' in line:
        continue
    else: # add file to all associated directories
        filesize = line.rpartition(' ')[0]
        temp_dir = cur_dir
        while '/' in temp_dir:
            if line[0:3] == 'dir':
                temp_dir = ''
            elif temp_dir in directories:
                directories[temp_dir] += int(filesize)
            else:
                directories[temp_dir] = int(filesize)
            temp_dir = temp_dir.rpartition('/')[0]
        
#print(directories)
    
```

```python
# part 1

dir_sizes = list(directories.values())

result = 0
for size in dir_sizes:
    if size <= 100000:
        result = result + size
        
print(result)
```

```python
for directory, size in directories.items():
    print('{} {}'.format(directory, size))
```

```python
# part 2

total_available = 70000000
need_unused =     30000000

need_to_free = directories['/home'] - (total_available - need_unused)

result = total_available
for size in dir_sizes:
    if size >= need_to_free:
        if size < result:
            result = size

print(result)
```
